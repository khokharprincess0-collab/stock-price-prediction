```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data load karein
df = pd.read_csv('Book1.csv')

# Missing values ko handle karein (agar koi hain)
df = df.dropna()

# X ko hum index (din) aur y ko 'Closing Volume' rakh rahe hain
X = np.arange(len(df)).reshape(-1, 1)
y = df['Closing Volume'].values
```

```python
# Linear Regression model banayein aur train karein
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Predictions
y_pred_lin = lin_reg.predict(X)

# Graph
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color='lightgray', label='Actual Stock Data', alpha=0.5)
plt.plot(X, y_pred_lin, color='blue', linewidth=2, label='Linear Regression')
plt.title('Model 1: Linear Regression')
plt.xlabel('Days')
plt.ylabel('Closing Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Polynomial features banayein (degree 3 ek accha starting point hai)
poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)

# Polynomial model train karein
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# Predictions
y_pred_poly = poly_reg.predict(X_poly)

# Graph
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color='lightgray', label='Actual Stock Data', alpha=0.5)
plt.plot(X, y_pred_poly, color='red', linewidth=2, label='Polynomial Regression')
plt.title('Model 2: Polynomial Regression')
plt.xlabel('Days')
plt.ylabel('Closing Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Asal data points
plt.scatter(X, y, color='lightgray', label='Actual Stock Data', alpha=0.3)

# Dono models ki predictions plot karein
plt.plot(X, y_pred_lin, color='blue', linewidth=2, label='Linear Regression (Blue)')
plt.plot(X, y_pred_poly, color='red', linewidth=2, label='Polynomial Regression (Red)')

# Graph ki formatting
plt.title('Linear vs Polynomial Regression on Stock Data')
plt.xlabel('Days')
plt.ylabel('Closing Volume')
plt.legend()
plt.grid(True)
plt.show()
```
